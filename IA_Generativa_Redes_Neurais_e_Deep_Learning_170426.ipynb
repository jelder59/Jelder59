```python
import tensorflow as tf
from tensorflow import keras
import numpy as np # Para biblioteca para dados
import matplotlib.pyplot as plt

print("Versão do TensorFlow: ", tf.__version__) # Usando dois underlines __ (Sintaxe da linguagem)
```

```python
fashion_mnist = keras.datasets.fashion_mnist
(train_images, train_labels), (test_images, test_labels) = fashion_mnist.load_data()

class_names = ['Camiseta','Calça','Sueter','Vestido','Casaco',
              'Sandália','Camisa','Tenis', 'Bolsa','Bota']

print(f"Imagens do Treino: {train_images.shape}")
print(f"Imagens do Teste: {test_images.shape}")
```

```python
plt.figure()
plt.imshow(train_images[-0], cmap = plt.cm.binary)
plt.colorbar()
plt.grid(False)
plt.title(f"Classe: {class_names[train_labels[-0]]}")
plt.show()
```

Normalização

```python
train_images = train_images / 255.0
test_images = test_images / 255.0
```

Construindo a MLP

```python
#Constuindo a MLP
modelo = keras.Sequential([
    keras.layers.Flatten(input_shape=(28,28)),# Camada de Entrada
    keras.layers.Dense(128, activation='relu'),# Camada Oculta (Aprendizado)
    keras.layers.Dense(10, activation='softmax')# Camada de Saída (Previsão)
])

modelo.summary()
```
